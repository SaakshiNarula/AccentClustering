```
import pandas as pd
Q = pd.read_excel('stats250_problemroulette_Q.xlsx')
A = pd.read_excel('stats250_problemroulette_ANS.xlsx')
Q_numbers = pd.read_csv('stats250_problemroulette_question_numbers.csv')
Chapter = pd.read_csv('stats250_problemroulette_chapter.csv')
```

```
#chapter 4 Q numbers
ref_id = int(Chapter[Chapter['order'] == 4]['id'])
q_numbers = list(Q_numbers[Q_numbers['topic_id'] == ref_id]['question_id'])
q_numbers
```

```
chp4_questions = Q[Q['id'].isin(q_numbers)]
```

```
updated_dict = {}
for (key, value) in dictOfWords.items():
  if value != 0:
    updated_dict[key] =value
updated_dict_keys = [*updated_dict]
```

```
updated_dict_keys
bank = pd.DataFrame()
for i in range(len(chp4_questions)):
  for key in updated_dict_keys:
    if key in chp4_questions.iloc[i]['text']:
     
      bank = bank.append(chp4_questions.iloc[i])
      break
bank
```
